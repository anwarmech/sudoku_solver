### Sudoku Solver

```python
import numpy as np
import tkinter as tk
from tkinter import *
from tkinter import ttk

window = tk.Tk()
window.title("Sudoku Solver")

lab = tk.Label(window,text="Enter Sudoku Numbers",font=25)
lab.grid(row=0,column=0,columnspan=9,pady=5)
global renew_button
renew_button = 0
text = []
a=0
b=0
c=0

#Grid Properties
h=1 
w=2 
ft=30 
thick=1 

# Generate Textbox Grid
for i in range(9):
    for j in range(9):
        c+=1
        if a<3:
            a+=1
            textbox = tk.Text(window,bg="lightgray",height=h, width=w,highlightthickness=thick,font=ft,padx=10,pady=10)
            textbox.grid(row=i+1,column=j)
            text.append(textbox)
            if c%9==0:
                a-=3
            if c%27==0:
                a=3
        elif a<6:
            a+=1
            textbox = tk.Text(window,bg="white",height=h, width=w,highlightthickness=thick,font=ft,padx=10,pady=10)
            textbox.grid(row=i+1,column=j)
            text.append(textbox)
            if a==6:
                a=0
                if c%9==0:
                    a=3 
                if c%54==0:
                    a=0
                    
### Function to solve sudoku
def capture():
    global renew_button
    if renew_button == 0:
        button1.destroy()
    else:
        rbt.grid_forget()
    renew_button = 0
    textvalues = []
    lab.config(text="Numbers Entered",fg='black')
    lab1 = tk.Label(window,text="Sudoku Solution",font=20)
    lab1.grid(row=0,column=10,columnspan=9)
    lab2 = tk.Label(window,text="",font=20,height=1, width=1)
    lab2.grid(row=0,column=9)
    
    # Creating Sudoku Array
    for i in text:
        value = i.get("1.0", "end-1c")
        textvalues.append(value)
    
    a = []
    for i in range(len(textvalues)):
        if textvalues[i] == '':
            a.append(0)
        else:
            a.append(int(textvalues[i]))
    arr = np.array(a).reshape(9,9) ## Initial Array
    arr1 = np.array(a).reshape(9,9) ## Array to capture missing numbers

    # Finding Missing Numbers in Array
    def calculate(arr1):
        zero={0}
        l=0
        for i in range(9):
            for j in range(9):
                x1 = l*3
                x2 = (l+1)*3
                y1 = (j//3)*3
                y2 = ((j//3)+1)*3
                if j==8 and (i+1)%3==0 and i>0:
                    l+=1
                if arr1[i,j]==0:
                    group = set((arr1[x1:x2,y1:y2]).flatten())-zero
                    row = set(arr1[i:i+1].flatten())-zero
                    column = set(arr1[:,j:j+1].flatten())-zero
                    present = group.union(row,column)
                    for n in range(9):
                        num=n+1
                        if num not in present:
                            arr1[i,j]=num
                            
                            if calculate(arr1):
                                return True
                            arr1[i,j]=0
                    return False
        return True

    calculate(arr1)
    
    ### Display Sudoku Result
    a=0
    b=0
    c=0
    for i in range(9):
        for j in range(9):
            c+=1
            if a<3:
                a+=1
                if arr[i,j]!=0:
                    label = tk.Label(window,text=arr[i,j],height=h, width=w,fg='red',bg="lightgray",highlightthickness=thick,font=ft,padx=10,pady=10)
                    label.grid(row=i+1,column=j+10)
                else:
                    label = tk.Label(window,text=arr1[i,j],height=h, width=w,fg='darkgreen',bg="lightgray",highlightthickness=thick,font=ft,padx=10,pady=10)
                    label.grid(row=i+1,column=j+10)                
                if c%9==0:
                    a-=3
                if c%27==0:
                    a=3 
            elif a<6:
                a+=1
                if arr[i,j]!=0:
                    label = tk.Label(window,text=arr[i,j],height=h, width=w,fg='red',bg="white",highlightthickness=thick,font=ft,padx=10,pady=10)
                    label.grid(row=i+1,column=j+10)
                else:
                    label = tk.Label(window,text=arr1[i,j],height=h, width=w,fg='darkgreen',bg="white",highlightthickness=thick,font=ft,padx=10,pady=10)
                    label.grid(row=i+1,column=j+10)
                if a==6:
                    a=0
                    if c%9==0:
                        a=3   
                    if c%54==0:
                        a=0
    ### Inserting Close Button
    def close():
        window.destroy()        
    button3 = tk.Button(window,text="Close",command=close,font=20,fg='Red')
    button3.grid(row=10,column=11,columnspan=3,pady=10)

    ### Defining function to Solve Sudoku Again 
    def renew():
        global renew_button
        renew_button+=1
        #global arr1
        #del capture.arr1
        button_3.destroy()
        global rbt
        rbt = tk.Button(window, text="Solve", fg='Blue',font=20,command=capture1)
        rbt.grid(row=10,column=3,columnspan=3,pady=10)
        reject()
        lab.config(text="Enter Sudoku Numbers",fg='black')
        return rbt

    ### Inserting Solve Again Button
    button_3 = tk.Button(window,text="Solve Again",command=renew,font=20,fg='Blue')
    button_3.grid(row=10,column=14,columnspan=3,pady=10)


# Function to checck repeated numbers
def num_scan(data_):
    zero = {0}
    row_no = 0
    global flag
    flag = 0
    for i in range(3):
        for j in range (3):
            def filter(data):
                a1 = [x for x in data if x!=0]
                a2 = set(data)-zero
                if len(a1) != len(a2):
                    global flag
                    flag+=1
            p1=i*3
            p2=(i+1)*3
            q1=j*3
            q2=(j+1)*3
            group_ = list(data_[p1:p2,q1:q2].flatten())
            filter(group_)
            row_ = data_[row_no]
            filter(row_)
            column_ = data_[:,row_no]
            filter(column_)
            row_no+=1
    if flag==0:
        return True
    else:
        return False

# Function to check incorrect entry
def capture1():
    all_num = ['','1','2','3','4','5','6','7','8','9']
    valid_num = ['1','2','3','4','5','6','7','8','9']
    min_num = 0
    int_num = 0
    temp_ = []
    temp_.clear()
    for i in text:
        value = i.get("1.0", "end-1c")
        if value == "":
            temp_.append(0)
        if value not in all_num:
            temp_.append(0)
            int_num+=1
        elif value in valid_num:
            temp_.append(int(value))
            min_num+=1
    temp_arr = np.array(temp_).reshape(9,9)
    if int_num==0 and min_num>0 and num_scan(temp_arr):
        capture()
    else:
        reject()
        global reject_num
        reject_num+=1

# Function when incorrect details submitted to solve sudoku
reject_num=1
def reject():     
    global reject_num
    rej = reject_num
    reject_text = 'Invalid numbers, try again ('+str(rej)+')....'
    lab.config(text=reject_text,fg="red")
    rej+=1 # Finding No of Times data entered incorrectly
    text.clear() # Deleting old data
    a=0
    b=0
    c=0

    # getting Newly Entered Data to Solve
    for i in range(9):
        for j in range(9):
            c+=1
            if a<3:
                a+=1
                textbox = tk.Text(window,bg="lightgray",height=h, width=w,highlightthickness=thick,font=ft,padx=10,pady=10)
                textbox.grid(row=i+1,column=j)
                text.append(textbox)
                if c%9==0:
                    a-=3
                if c%27==0:
                    a=3
            elif a<6:
                a+=1
                textbox = tk.Text(window,bg="white",height=h, width=w,highlightthickness=thick,font=ft,padx=10,pady=10)
                textbox.grid(row=i+1,column=j)
                text.append(textbox)
                if a==6:
                    a=0
                    if c%9==0:
                        a=3   
                    if c%54==0:
                        a=0

## Inserting Solve Button
button1 = tk.Button(window, text="Solve", fg='Blue',font=20,command=capture1)
button1.grid(row=10,column=3,columnspan=3,pady=10)

window.mainloop()
```
